## Solução do Desafio: Calculadora de Troco

```python
# Solução básica
print("=== Caixa Registradora ===")
print()

preco = float(input("Preço do produto (R$): "))
valor_pago = float(input("Valor pago pelo cliente (R$): "))

troco = valor_pago - preco

print()
print(f"Preço do produto: R$ {preco:.2f}")
print(f"Valor pago: R$ {valor_pago:.2f}")
print(f"Troco: R$ {troco:.2f}")
```

```python
# Solução avançada: verifica se o valor pago é suficiente
print("=== Caixa Registradora ===")
print()

preco = float(input("Preço do produto (R$): "))
valor_pago = float(input("Valor pago pelo cliente (R$): "))

print()
print(f"Preço: R$ {preco:.2f}")
print(f"Pago: R$ {valor_pago:.2f}")

if valor_pago >= preco:
    troco = valor_pago - preco
    if troco == 0:
        print("Pagamento exato, sem troco!")
    else:
        print(f"Troco: R$ {troco:.2f}")
else:
    falta = preco - valor_pago
    print(f"Valor insuficiente! Faltam R$ {falta:.2f}")
```
