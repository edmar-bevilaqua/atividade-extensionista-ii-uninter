## Solução do Desafio: FizzBuzz

```python
# FizzBuzz de 1 a 30
# ATENÇÃO: verificar FizzBuzz ANTES de Fizz e Buzz separados!

for numero in range(1, 31):
    if numero % 3 == 0 and numero % 5 == 0:
        print("FizzBuzz")
    elif numero % 3 == 0:
        print("Fizz")
    elif numero % 5 == 0:
        print("Buzz")
    else:
        print(numero)
```

```python
# Versão alternativa: imprime tudo na mesma linha separado por vírgula
resultados = []

for numero in range(1, 31):
    if numero % 15 == 0: # 15 = mmc(3, 5)
        resultados.append("FizzBuzz")
    elif numero % 3 == 0:
        resultados.append("Fizz")
    elif numero % 5 == 0:
        resultados.append("Buzz")
    else:
        resultados.append(str(numero))

print(", ".join(resultados))
```
