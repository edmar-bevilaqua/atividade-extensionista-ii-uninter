## Parte 1: O Módulo `random`

O módulo `random` permite gerar números **aleatórios**, como se o computador jogasse um dado!

Para usar, precisamos **importar** o módulo com `import random`.

````python
import random

# randint(a, b) gera um número inteiro entre a e b (inclusive)
numero = random.randint(1, 10)
print(f"Número aleatório entre 1 e 10: {numero}")

# Execute várias vezes: o número muda!
for i in range(5):
    dado = random.randint(1, 6)
    print(f"Jogada {i+1}: {dado}")
````

## Parte 2: Listas: Guardando Vários Valores

Uma **lista** é como uma caixa que guarda **vários itens** em ordem.

```python
frutas = ["maçã", "banana", "laranja"]
# 0 1 2 ← posições (índices)
```

Vamos usar uma lista para guardar os **palpites do jogador** no nosso jogo!

````python
# Criando e usando listas
cores = ["vermelho", "azul", "verde", "amarelo"]

print(f"A lista tem {len(cores)} cores") # len() conta os itens
print(f"Primeira cor: {cores[0]}") # índice 0 = primeiro item
print(f"Última cor: {cores[-1]}") # índice -1 = último item

# Percorrendo com for
for cor in cores:
    print(f" - {cor}")
````

````python
# Adicionando itens à lista com .append()
palpites = [] # lista vazia

palpites.append(15)
palpites.append(42)
palpites.append(7)

print(f"Palpites feitos: {palpites}")
print(f"Total de palpites: {len(palpites)}")
````

## Construindo o Jogo: Passo a Passo!

Vamos construir o jogo **incrementalmente**, peça por peça.

### Passo 1: O computador pensa em um número

````python
import random

# O computador escolhe um número secreto entre 1 e 100
numero_secreto = random.randint(1, 100)

print("=== JOGO DE ADIVINHAÇÃO ===")
print("Pensei em um número entre 1 e 100!")
print(f"(psiu: o número é {numero_secreto}, só para você ver agora)")
````

### Passo 2: O jogador faz um palpite

````python
# O jogador faz um palpite e o programa dá uma dica
palpite = int(input("Qual é o seu palpite? "))

if palpite < numero_secreto:
    print(" O número é MAIOR que o seu palpite!")
elif palpite > numero_secreto:
    print(" O número é MENOR que o seu palpite!")
else:
    print(" ACERTOU!")
````

### Passo 3: Permitir múltiplas tentativas com `while`

````python
import random

numero_secreto = random.randint(1, 100)
tentativas = 0
acertou = False

print("=== JOGO DE ADIVINHAÇÃO ===")
print("Pensei em um número entre 1 e 100. Pode tentar!")
print()

while not acertou:
    palpite = int(input("Seu palpite: "))
    tentativas = tentativas + 1
    
    if palpite < numero_secreto:
        print(" Mais alto!")
    elif palpite > numero_secreto:
        print(" Mais baixo!")
    else:
        acertou = True

print()
print(f" PARABÉNS! Você acertou em {tentativas} tentativa(s)!")
````

### Passo 4: Limitar o número de tentativas

````python
import random

numero_secreto = random.randint(1, 100)
tentativas = 0
max_tentativas = 7 # máximo de tentativas permitidas
acertou = False

print("=== JOGO DE ADIVINHAÇÃO ===")
print(f"Pensei em um número entre 1 e 100. Você tem {max_tentativas} tentativas!")
print()

while tentativas < max_tentativas and not acertou:
    restantes = max_tentativas - tentativas
    palpite = int(input(f"[{restantes} tentativas restantes] Seu palpite: "))
    tentativas = tentativas + 1
    
    if palpite < numero_secreto:
        print(" Mais alto!")
    elif palpite > numero_secreto:
        print(" Mais baixo!")
    else:
        acertou = True

print()

if acertou:
    print(f" PARABÉNS! Você acertou em {tentativas} tentativa(s)!")
else:
    print(f" Não foi dessa vez. O número era {numero_secreto}.")
````

### Passo 5 (Bônus): Guardar palpites e adicionar opção de jogar novamente

````python
import random

def jogar():
    numero_secreto = random.randint(1, 100)
    tentativas = 0
    max_tentativas = 7
    acertou = False
    historico = [] # lista para guardar os palpites

    print()
    print("=" * 35)
    print(" JOGO DE ADIVINHAÇÃO ")
    print("=" * 35)
    print(f"Pensei em um número entre 1 e 100.")
    print(f"Você tem {max_tentativas} tentativas. Boa sorte!")
    print()

    while tentativas < max_tentativas and not acertou:
        restantes = max_tentativas - tentativas
        palpite = int(input(f"[{restantes} restantes] Seu palpite: "))
        tentativas += 1
        historico.append(palpite) # guarda o palpite na lista

        if palpite < numero_secreto:
            print(" Mais alto!\n")
        elif palpite > numero_secreto:
            print(" Mais baixo!\n")
        else:
            acertou = True

    print()
    print(f"Seus palpites foram: {historico}")
    print()

    if acertou:
        if tentativas == 1:
            print("🏆 INCRÍVEL! Acertou de PRIMEIRA!")
        elif tentativas <= 3:
            print(f" EXCELENTE! Acertou em {tentativas} tentativas!")
        else:
            print(f" Parabéns! Acertou em {tentativas} tentativas!")
    else:
        print(f" Não foi dessa vez. O número era {numero_secreto}.")

# Loop principal do jogo
jogar()

while True:
    resposta = input("\nQuer jogar novamente? (s/n) ").lower()
    if resposta == "s":
        jogar()
    else:
        print("\nObrigado por jogar! Até a próxima! ")
        break
````
