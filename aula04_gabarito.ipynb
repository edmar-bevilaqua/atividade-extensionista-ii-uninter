### Jogo de Adivinhação: Versão Completa

> Este é o jogo completo desenvolvido ao longo da aula, com todas as funcionalidades implementadas.

```python
import random

def calcular_pontuacao(tentativas, max_tentativas):
    """Calcula a pontuação com base no número de tentativas usadas."""
    pontuacao = (max_tentativas - tentativas + 1) * 100
    return max(pontuacao, 0)

def jogar(dificuldade):
    """Executa uma rodada do jogo com a dificuldade escolhida."""
    configuracoes = {
        "1": {"max": 50, "tentativas": 10, "nome": "Fácil"},
        "2": {"max": 100, "tentativas": 7, "nome": "Normal"},
        "3": {"max": 200, "tentativas": 5, "nome": "Difícil"},
    }

    config = configuracoes[dificuldade]
    numero_secreto = random.randint(1, config["max"])
    max_tentativas = config["tentativas"]
    tentativas = 0
    acertou = False
    historico = []

    print()
    print("=" * 40)
    print(f" JOGO DE ADIVINHAÇÃO, {config['nome'].upper()}")
    print("=" * 40)
    print(f"Pensei em um número entre 1 e {config['max']}.")
    print(f"Você tem {max_tentativas} tentativas. Boa sorte!")
    print()

    while tentativas < max_tentativas and not acertou:
        restantes = max_tentativas - tentativas

        try:
            palpite = int(input(f"[{restantes} restantes] Seu palpite (1-{config['max']}): "))
        except ValueError:
            print(" Por favor, digite um número válido!")
            continue

        if palpite < 1 or palpite > config["max"]:
            print(f" O número deve estar entre 1 e {config['max']}!")
            continue

        tentativas += 1
        historico.append(palpite)

        if palpite < numero_secreto:
            print(" Mais alto!\n")
        elif palpite > numero_secreto:
            print(" Mais baixo!\n")
        else:
            acertou = True

    print()
    print(f"Seus palpites: {historico}")
    print()

    if acertou:
        pontuacao = calcular_pontuacao(tentativas, max_tentativas)
        if tentativas == 1:
            print("🏆 INCRÍVEL! Acertou de PRIMEIRA! Você é um gênio!")
        elif tentativas <= 3:
            print(f" EXCELENTE! Acertou em {tentativas} tentativas!")
        else:
            print(f" Parabéns! Acertou em {tentativas} tentativas!")
        print(f"⭐ Sua pontuação: {pontuacao} pontos")
        return pontuacao
    else:
        print(f" Não foi dessa vez. O número era {numero_secreto}.")
        return 0


# ─── TELA INICIAL ───────────────────────────────────────
print()
print("╔══════════════════════════════════╗")
print("║ 🐍 JOGO DE ADIVINHAÇÃO 🐍 ║")
print("║ Criado com Python, Aula 04 ║")
print("╚══════════════════════════════════╝")

pontuacao_total = 0
rodadas = 0

while True:
    print()
    print("Escolha a dificuldade:")
    print(" 1 - Fácil (1-50, 10 tentativas)")
    print(" 2 - Normal (1-100, 7 tentativas)")
    print(" 3 - Difícil (1-200, 5 tentativas)")
    print(" 0 - Sair")
    print()

    escolha = input("Sua escolha: ")

    if escolha == "0":
        print()
        print(f"Você jogou {rodadas} rodada(s).")
        print(f"Pontuação total: {pontuacao_total} pontos")
        print()
        print("Obrigado por jogar! Até a próxima! ")
        break
    elif escolha in ["1", "2", "3"]:
        pontuacao = jogar(escolha)
        pontuacao_total += pontuacao
        rodadas += 1
    else:
        print("Opção inválida. Escolha 0, 1, 2 ou 3.")
```
