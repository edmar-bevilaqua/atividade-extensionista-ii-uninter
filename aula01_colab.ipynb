## Parte 2: Olá, Mundo!

A tradição em programação é que o primeiro programa que qualquer pessoa escreve exibe a mensagem **"Olá, Mundo!"**.

Em Python, usamos o comando `print()` para mostrar texto na tela.

**Regra:** o texto precisa estar entre aspas `"..."` dentro dos parênteses.

````python
# Meu primeiro programa em Python!
print("Olá, Mundo!")
````

````python
# Podemos usar print() várias vezes
# Cada print() cria uma linha nova na saída

print("Meu nome é Python.")
print("Tenho 33 anos de idade.")
print("Sou uma linguagem de programação!")
````

---
## Parte 3: Comentários

Comentários são mensagens que **você escreve para humanos**, não para o computador.

- Começam com `#`
- O Python ignora tudo que vem depois do `#` na mesma linha
- São usados para explicar o que o código faz

**Boa prática:** sempre comente seu código para que outras pessoas (e você no futuro!) entendam o que foi feito.

````python
# Este comentário não aparece na saída
print("Esta linha aparece!") # comentário no final da linha

# print("Esta linha está 'comentada', então não executa")

print("Apenas as linhas com print() aparecem na saída.")
````

Erros fazem parte da programação. **Todo programador** comete erros, o importante é saber lê-los!

Vamos provocar um erro de propósito para aprender a interpretá-lo.

````python
# ATENÇÃO: esta célula vai gerar um erro de propósito!
# Observe a mensagem que aparece e tente entender o que ela diz.

print("Olá, esqueci de fechar as aspas...)
````

O erro **`SyntaxError`** significa que o Python não conseguiu entender o código porque tem algo escrito errado, como uma palavra mal escrita para uma pessoa.

Agora veja a versão corrigida:

````python
# Versão corrigida: as aspas estão fechadas!
print("Olá, agora funcionou!")
````

## Parte 5: Arte com Caracteres (ASCII Art)

Você sabia que dá para desenhar usando apenas `print()`?
Esse tipo de arte com caracteres se chama **ASCII Art**.

Veja o exemplo abaixo:

````python
# Desenhando uma casa com print()
print(" /\\ ")
print(" / \\ ")
print(" / \\ ")
print(" /______\ ")
print(" | | ")
print(" | [==] | ")
print(" |______| ")
````

````python
# Desenhando um rosto feliz
print(" O O ")
print(" ")
print(" \\___/ ")
````

````python
# Usando print() com linha em branco
# print() sem nada dentro cria uma linha vazia

print("Linha 1")
print() # linha em branco
print("Linha 3")
````
