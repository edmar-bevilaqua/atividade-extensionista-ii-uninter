## Parte 1: Variáveis: Caixas com Etiquetas

Imagine que variáveis são **caixas** onde você guarda informações, e cada caixa tem uma **etiqueta** com o nome.

```
  ┌──────────┐
  │ "Maria" │ ← conteúdo da caixa
  └──────────┘
      nome ← etiqueta (nome da variável)
```

Para criar uma variável em Python, usamos o sinal de `=`:
```python
nome = "Maria"
```

````python
# Criando variáveis
nome = "Ana"
idade = 13
altura = 1.65

# Usando as variáveis com print()
print(nome)
print(idade)
print(altura)
````

````python
# Você pode mudar o valor de uma variável a qualquer momento
pontos = 0
print("Pontos iniciais:", pontos)

pontos = 10
print("Depois de ganhar pontos:", pontos)

pontos = pontos + 5 # somar 5 ao valor atual
print("Depois de mais 5 pontos:", pontos)
````

## Parte 2: Tipos de Dados

Em Python, cada valor tem um **tipo**. Os principais são:

| Tipo | Nome em Python | Exemplos |
|------|----------------|----------|
| Número inteiro | `int` | `5`, `100`, `-3` |
| Número decimal | `float` | `3.14`, `1.65`, `-0.5` |
| Texto | `str` | `"Olá"`, `"Maria"`, `"123"` |
| Verdadeiro/Falso | `bool` | `True`, `False` |

````python
# Exemplos de cada tipo
numero_inteiro = 42
numero_decimal = 3.14
texto = "Olá, Python!"
verdadeiro_ou_falso = True

# type() mostra o tipo de uma variável
print(type(numero_inteiro))
print(type(numero_decimal))
print(type(texto))
print(type(verdadeiro_ou_falso))
````

````python
# ATENÇÃO: "123" (str) é diferente de 123 (int)!
numero = 123
texto_numero = "123"

print(type(numero)) # int
print(type(texto_numero)) # str

# Com int podemos fazer contas, com str não!
print(numero + 1) # 124
print(texto_numero + "1") # "1231", junta os textos!
````

## Parte 3: Operações Matemáticas

| Operador | Operação | Exemplo | Resultado |
|----------|----------|---------|----------|
| `+` | Adição | `5 + 3` | `8` |
| `-` | Subtração | `10 - 4` | `6` |
| `*` | Multiplicação | `3 * 4` | `12` |
| `/` | Divisão | `10 / 4` | `2.5` |
| `//` | Divisão inteira | `10 // 4` | `2` |
| `%` | Resto da divisão | `10 % 3` | `1` |
| `**` | Potenciação | `2 ** 8` | `256` |

````python
# Calculadora simples
a = 15
b = 4

print("Adição:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Divisão inteira:", a // b)
print("Resto:", a % b)
print("Potência (a ao cubo):", a ** 3)
````

````python
# O operador % (módulo) é muito útil!
# Ele retorna o RESTO da divisão

# Para saber se um número é par: resto da divisão por 2 é zero
numero = 17
print(numero, "% 2 =", numero % 2) # se for 0, é par; se for 1, é ímpar
````

## Parte 4: Interagindo com o Usuário: `input()`

O comando `input()` **pausa o programa** e espera o usuário digitar algo.
O que o usuário digitar é guardado em uma variável.

**Importante:** `input()` sempre retorna um **texto** (`str`), mesmo se o usuário digitar um número!

````python
# Pedindo o nome do usuário
nome = input("Qual é o seu nome? ")
print("Olá,", nome, "!")
````

````python
# Para usar um número digitado pelo usuário,
# precisamos converter com int() ou float()

idade_texto = input("Qual é a sua idade? ")
idade = int(idade_texto) # converte de str para int

print("Você tem", idade, "anos.")
print("No ano que vem você terá", idade + 1, "anos!")
````

````python
# Atalho: converter direto na mesma linha
altura = float(input("Qual é a sua altura em metros? (ex: 1.65) "))
print("Sua altura é", altura, "metros.")
````

## Parte 5: Formatando Texto com f-strings

As **f-strings** são a forma mais moderna e elegante de misturar variáveis com texto.

Coloque um `f` antes das aspas e use `{nome_da_variavel}` dentro do texto:

````python
nome = "Carlos"
idade = 14
cidade = "Guaíba"

# Sem f-string (mais trabalhoso)
print("Olá, " + nome + "! Você tem " + str(idade) + " anos e mora em " + cidade + ".")

# Com f-string (muito mais fácil!)
print(f"Olá, {nome}! Você tem {idade} anos e mora em {cidade}.")
````

````python
# f-strings também permitem fazer cálculos dentro das chaves!
preco = 25.50
quantidade = 3

print(f"Você comprou {quantidade} itens a R${preco:.2f} cada.")
print(f"Total: R${preco * quantidade:.2f}")

# O :.2f arredonda para 2 casas decimais
````

## Projeto: Calculadora de IMC

Vamos criar uma calculadora de **Índice de Massa Corporal (IMC)**!

**Fórmula:** `IMC = peso / (altura * altura)`

O programa deve:
1. Pedir o nome, peso e altura do usuário
2. Calcular o IMC
3. Exibir o resultado formatado

````python
# Calculadora de IMC
print("=== Calculadora de IMC ===")
print()

# Pedir dados ao usuário
nome = input("Qual é o seu nome? ")
peso = float(input("Qual é o seu peso em kg? (ex: 65.5) "))
altura = float(input("Qual é a sua altura em metros? (ex: 1.70) "))

# Calcular o IMC
imc = peso / (altura ** 2)

# Mostrar o resultado
print()
print(f"Olá, {nome}!")
print(f"Seu IMC é: {imc:.1f}")
print()
print("Tabela de referência:")
print(" Abaixo de 18.5 → Abaixo do peso")
print(" 18.5 a 24.9 → Peso normal")
print(" 25.0 a 29.9 → Sobrepeso")
print(" 30.0 ou mais → Obesidade")
````
