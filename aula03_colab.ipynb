## Parte 1: Operadores de Comparação

Antes de aprender `if`, precisamos saber como **comparar** valores.
O resultado de uma comparação é sempre `True` ou `False`.

| Operador | Significado | Exemplo | Resultado |
|----------|-------------|---------|----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Diferente de | `5 != 3` | `True` |
| `>` | Maior que | `10 > 7` | `True` |
| `<` | Menor que | `3 < 1` | `False` |
| `>=` | Maior ou igual | `5 >= 5` | `True` |
| `<=` | Menor ou igual | `4 <= 3` | `False` |

````python
# Testando comparações
print(10 > 5) # True
print(3 == 4) # False
print(7 != 7) # False
print("ana" == "ana") # True, funciona com texto também!
````

## Parte 2: Condicionais: `if`, `elif`, `else`

O `if` permite que o programa **tome decisões**.

```
SE (condição for verdadeira):
    execute este bloco
SENÃO SE (outra condição):
    execute este bloco
SENÃO:
    execute este bloco
```

> O bloco de código dentro do `if` **precisa ter 4 espaços** de indentação!

````python
# Verificando se um número é positivo, negativo ou zero
numero = int(input("Digite um número: "))

if numero > 0:
    print("O número é positivo! ✅")
elif numero < 0:
    print("O número é negativo! ")
else:
    print("O número é zero! ")
````

````python
# Classificação de faixa etária
idade = int(input("Qual é a sua idade? "))

if idade < 12:
    categoria = "Criança"
elif idade < 18:
    categoria = "Adolescente"
elif idade < 60:
    categoria = "Adulto"
else:
    categoria = "Idoso"

print(f"Com {idade} anos, você é: {categoria}")
````

````python
# Combinando condições com 'and' e 'or'
nota = float(input("Qual foi sua nota? (0 a 10) "))

if nota >= 7.0:
    print("Aprovado! ")
elif nota >= 5.0 and nota < 7.0:
    print("Recuperação ")
else:
    print("Reprovado ")
````

## Parte 3: Laço `for`: Repetindo um Número Fixo de Vezes

O `for` é perfeito quando você sabe **quantas vezes** quer repetir algo.

A função `range(n)` gera números de `0` até `n-1`:
- `range(5)` → `0, 1, 2, 3, 4`
- `range(1, 6)` → `1, 2, 3, 4, 5`
- `range(0, 10, 2)` → `0, 2, 4, 6, 8`

````python
# Contando de 1 a 5
for numero in range(1, 6):
    print(numero)
````

````python
# Tabuada do 7
numero = 7
print(f"=== Tabuada do {numero} ===")

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2} = {resultado}")
````

````python
# Desenhando uma pirâmide com for!
tamanho = 5

for i in range(1, tamanho + 1):
    print("*" * i)
````

````python
# Percorrendo uma lista de itens com for
frutas = ["maçã", "banana", "laranja", "uva"]

for fruta in frutas:
    print(f"Fruta: {fruta}")
````

## Parte 4: Laço `while`: Repetindo Enquanto uma Condição for Verdadeira

O `while` repete **enquanto** uma condição for `True`.
Útil quando você **não sabe** quantas vezes vai precisar repetir.

> Cuidado com **loops infinitos**: se a condição nunca se tornar `False`, o programa nunca para!

````python
# Contagem regressiva
contagem = 10

while contagem > 0:
    print(contagem)
    contagem = contagem - 1 # diminui 1 a cada repetição

print(" LANÇAMENTO!")
````

````python
# while com break: sai do laço quando uma condição especial é atingida
print("Digite números para somar. Digite 0 para parar.")

total = 0

while True: # laço que parece infinito...
    numero = int(input("Número: "))
    
    if numero == 0:
        break # ...mas o break para tudo!
    
    total = total + numero

print(f"Total: {total}")
````

## Mini-Jogo: Par ou Ímpar!

Vamos criar um jogo simples usando tudo que aprendemos!

**Regras:**
- O jogador escolhe um número de 1 a 10
- O computador também escolhe um número de 1 a 10
- Somamos os dois números
- O jogador aposta se a soma é par ou ímpar
- Mostramos o resultado!

````python
import random # módulo para gerar números aleatórios

print("=== PAR OU ÍMPAR! ===")
print()

# Jogador escolhe um número
numero_jogador = int(input("Escolha um número de 1 a 10: "))

# Computador escolhe um número aleatório
numero_computador = random.randint(1, 10)
print(f"O computador escolheu: {numero_computador}")

# Soma dos dois números
soma = numero_jogador + numero_computador
print(f"Soma: {numero_jogador} + {numero_computador} = {soma}")
print()

# Jogador aposta
aposta = input("É PAR ou ÍMPAR? Digite P ou I: ").upper()

# Verificar se a soma é par ou ímpar
if soma % 2 == 0:
    resultado = "P" # Par
else:
    resultado = "I" # Ímpar

# Comparar aposta com resultado
if aposta == resultado:
    print(" Você GANHOU!")
else:
    print(" Você perdeu. Tente de novo!")
````
